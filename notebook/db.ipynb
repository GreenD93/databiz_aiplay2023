```python
import pandas as pd

user_profile_items = {
    'id': ['aiplay2023', 'kt', 'kbank'],
    'pwd': ['123', '123', '123'],
    'name': ['aiplay2023', 'kt', 'kbank'],
    'geng': ['여성','남성','여성'],
    'ageg': ['20대','40대','30대'],
    'jobseg': ['학생','전문직','주부']
}

user_df = pd.DataFrame.from_dict(user_profile_items)
user_df.head()
```

```python
try:
    with db.cursor() as curs:
    
        curs.execute('SET SESSION TRANSACTION ISOLATION LEVEL READ UNCOMMITTED')
        db.commit()
        
        for row in user_df.iterrows():
            _, data = row
            
            id = data['id']
            pwd = data['pwd']
            name = data['name']
            geng = data['geng']
            ageg = data['ageg']
            jobseg = data['jobseg']
        
            sql = """
                        INSERT INTO user_profile (
                            id, pwd,
                            name, geng, ageg, jobseg
                        )
                        VALUES (
                            %s, %s,
                            %s, %s, %s, %s
                        );
                """    
        
            sql_args = (
                    id, pwd,
                    name, geng, ageg, jobseg
            )
        
            curs.execute(sql, sql_args)
            
except Exception as e:
    print('error')
    raise e

db.commit()
```

## MKT PHRASE
